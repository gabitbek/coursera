```python
import pandas as pd
df_data = pd.read_csv ("fatalaccidentdata.csv") 

# block 1 - simple stats
sum1 = df_data['Pedal_Cycles'].sum()
count1 = df_data['Pedal_Cycles'].count()
count2 = sum1/count1*100


# print block 1
print ('Number of pedal cycle accidents: ' + str(sum1))
print ('Number of all accidents: ' + str(count1))
print ('Share of pedal cycle accidents: ' + str(round(count2)) + ' percents')
```
